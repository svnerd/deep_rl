```python
import numpy as np

def epsilon_greedy_prob(Q, s, epsilon=0.01):
    aN = len(Q[s])
    ai = np.argmax(Q[s])
    mainP = (1 - epsilon)
    c = epsilon / (aN - 1)
    p = [ mainP if i == ai else c for i in range(aN)]
    return p

def epsilon_greedy(Q, s, epsilon=0.01):
    p = epsilon_greedy_prob(Q, s, epsilon)
    return np.random.choice(len(p), 1, p=p)[0]

def sarsa_one_episode(env, Q, alpha):
    s0 = env.reset()
    a0 = epsilon_greedy(Q, s0)
    while True:
        s1, r, done, info = env.step(a0)
        if done:
            break
        a1 = epsilon_greedy(Q, s1)
        Q[s0][a0] = Q[s0][a0] + alpha * (r + Q[s1][a1] - Q[s0][a0])
        a0 = a1
        s0 = s1

def sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        
        ## TODO: complete the function
        sarsa_one_episode(env, Q, alpha)
    return Q

def q_learning_one_episode(env, Q, alpha):
    s0 = env.reset()
    while True:
        a0 = epsilon_greedy(Q, s0)
        s1, r, done, info = env.step(a0)
        if done:
            break
        m = np.max(Q[s1])
        Q[s0][a0] = Q[s0][a0] + alpha * (r + m - Q[s0][a0])
        s0 = s1


def q_learning(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        q_learning_one_episode(env, Q, alpha)
    return Q



def expected_sarsa_one_episode(env, Q, alpha):
    s0 = env.reset()
    while True:
        a0 = epsilon_greedy(Q, s0)
        s1, r, done, info = env.step(a0)
        if done:
            break
        p = epsilon_greedy_prob(Q, s1)
        e = np.dot(Q[s1], p)
        Q[s0][a0] = Q[s0][a0] + alpha * (r + e - Q[s0][a0])
        s0 = s1

def expected_sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        expected_sarsa_one_episode(env, Q, alpha)
    return Q

Q_sarsa = sarsa(env, 10000, 0.01)
Q_qlearning = q_learning(env, 10000, 0.01)
Q_expsarsa = expected_sarsa(env, 10000, 0.01)
```
